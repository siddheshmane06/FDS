# Iris dataset

```python
import pandas as pd
iris = pd.read_csv("iris.csv")
print(iris.head())
```

```python
#get columns to plot
columns = iris.columns.drop(["variety"])
#create x data
x_data = range(0, iris.shape[0])
#create figure and axis
fig, ax = plt.subplots()
#plot each column
for column in columns:
    ax.plot(x_data, iris[column],label = column)
    
#set title and legend
ax.set_title("Iris Dataset")
ax.legend()
```

```python
#create figure and axis
fig, ax = plt.subplots()
#plot histogram
ax.hist(iris["sepal.length"])
#set title and labels
ax.set_title("Iris")
ax.set_xlabel("Sepal_Length")
ax.set_ylabel("Frequency")
```

# Wine dataset

```python
wine_data = pd.read_csv("winemag.csv")
wine_data.head()
```

```python
#bar chart
#create a figure and axis
fig, ax = plt.subplots()
#count the occurence of each class
data= wine_data['points'].value_counts()
#get x and y data
points = data.index
frequency = data.values
#create bar charts
ax.bar(points,frequency)
#set title and label
ax.set_title("Wine Review Scores")
ax.set_xlabel("Points")
ax.set_ylabel("Frequency")
```
